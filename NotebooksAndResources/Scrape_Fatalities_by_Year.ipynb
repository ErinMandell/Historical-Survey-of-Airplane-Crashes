## Import Dependencies

```python
# Dependencies
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from sqlalchemy import create_engine
from pprint import pprint
```

## Create path to website, initialize browser, scrape page

```python
#launches a browser when headless = False
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#sends my browser to this url
url_baaa = 'http://www.baaa-acro.com/statistics/death-rate-per-year'
browser.visit(url_baaa)
```

```python
# HTML object
html_baaa = browser.html
    
# Parse HTML with Beautiful Soup
soup_baaa = bs(html_baaa, 'html.parser')
#soup_baaa
```

```python
# Examine the results, then determine element that contains sought info
print(soup_baaa.prettify())
```

## Grab data and assign to variables

```python
rankings = []
years = []
fatalitiesT = []
```

```python
result_rank = soup_baaa.find_all('td', class_='views-field views-field-counter')
# result_rank
```

```python
for rank in result_rank:
    ranking = rank.text
    rankings.append((ranking))
    
print(rankings)
```

```python
type(rankings)
```

```python
rankingsS = pd.Series(rankings)
rankingsS
```

## Convert data strings to series

```python
result_year = soup_baaa.find_all('td', class_='views-field views-field-created')

for year in result_year:
    year = year.text
    years.append((year))
    
print(years)
```

```python
yearsS = pd.Series(years)
yearsS
```

```python
result_fatalities = soup_baaa.find_all('td', class_='views-field views-field-field-final-total-fatalities')

for fatality in result_fatalities:
    totalfatalities = fatality.text
    fatalitiesT.append(totalfatalities)

print(fatalitiesT)
```

```python
fatalitiesS = pd.Series(fatalitiesT)
fatalitiesS
```

```python
browser.quit()
```

## Create Data frame from series, format data types

```python
df = pd.DataFrame(data = dict(Ranking = rankingsS, Year = yearsS, TotalFatalities = fatalitiesS), index = rankingsS.index)
```

```python
df.head()
```

```python
df['Ranking'] = pd.to_numeric(df['Ranking'])
df['TotalFatalities'] = pd.to_numeric(df['TotalFatalities'])
```

```python
df.dtypes
```

```python
df['Year'] = pd.to_numeric(df['Year'])
df.dtypes
```
